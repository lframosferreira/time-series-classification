```python
import numpy as np
import h5sparse
import pandas as pd

PATH: str = "../FingerMovements.hdf5"
DIMENSIONS: int = 28
SIZE: int = 316
```

```python
ranking: np.array = np.zeros((DIMENSIONS, DIMENSIONS), dtype=np.int_)
with h5sparse.File(PATH, "r") as f:
    less_contribute_lists = f["diversity_rank"]
    for i in range(len(less_contribute_lists)):
        ranking[less_contribute_lists[str(i)][()], np.arange(DIMENSIONS)] += 1
df_1 = pd.DataFrame((np.round(ranking / SIZE, 2)).tolist()).sort_values(by=DIMENSIONS-1, ascending=False)
```
