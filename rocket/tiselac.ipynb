```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import json
import pickle
```

```python
def plot_f1_scores(json_data: dict) -> None:
    print(f'Accuracy: {json_data["accuracy"]}')
    plt.title("tiselac")
    plt.xticks(np.arange(9))
    plt.xlabel("Classes dos exames")
    plt.ylabel("F1-score (%)")
    bar_rects = plt.bar(np.arange(9), json_data["f1_score"], linewidth=1.5, edgecolor="black", alpha=0.5, align="center")
    for rect in bar_rects:
        height = round(rect.get_height(), 2)
        plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')
    plt.show()
```

```python
with open("tiselac/all/run_info_04-24-2023_03:51:57.json", "r") as f:
    data = json.load(f)
    plot_f1_scores(data)
```

```python
with open("tiselac/7_8/run_info_04-25-2023_16:05:05.json", "r") as f:
    data = json.load(f)
    plot_f1_scores(data)
```
