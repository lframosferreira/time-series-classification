```python
import pandas as pd
import numpy as np
import os
import json
import glob
```

```python
path: str = "/home/lfrf/ic/time-series-classification/results"
aaa = []
with open("res.json", "a") as f:
    for d in os.listdir(path):
        for d1 in os.listdir(path + f"/{d}"):
            for d2 in os.listdir(path + f"/{d}" + f"/{d1}"):
                for jsf in glob.glob(path + f"/{d}" + f"/{d1}" + f"/{d2}/*.json"):
                    with open(jsf, "r") as f2:
                        aux = json.load(f2)
                        #json.dump(aux, f, indent=4)
                        aaa.append(aux)
    json.dump(aaa, f, indent=4)
```

```python
df = pd.read_json("res.json")
df["dimensions_used"].update(df.pop("dimensios_used"))
df
```

```python
import math

for i, row in df[df["classification_report"].notna()].iterrows():
    cr = row["classification_report"]
    del cr["macro avg"]
    del cr["weighted avg"]
    rec = []
    prec = []
    f1 = []
    acc = cr["accuracy"]
    del cr["accuracy"]
    for k, v in cr.items():
        rec.append(v["recall"])
        prec.append(v["precision"])
        f1.append(v["f1-score"])
    df.at[i, "accuracy"] = acc or []
    df.at[i, "recall"] = rec or []
    df.at[i, "precision"] = prec or []
    df.at[i, "f1_score"] = f1 or []
    
df.drop(columns=["classification_report"], inplace=True)
```
